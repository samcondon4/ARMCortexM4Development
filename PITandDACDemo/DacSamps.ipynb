```python
import thinkdsp
import numpy as np
import matplotlib.pyplot as plot
import math
```

```python
s = thinkdsp.Sinusoid(freq=300, amp=1.65)
```

```python
ts = np.linspace(0,3*s.period,10000)
```

```python
Sin = []
for k in ts:
    Sin.append(s.evaluate(k) + 1.65)
```

This is the desired Sin Wave we want to see output from the DAC

```python
plot.plot(ts, Sin)
```

The formula for the DAC ouput is DACo = Vref*((1+DAT[0])/2^12)
Therefore DAT[0] = {(2^12)[1.65(1+sin(600*pi*t)]}/3.3  -  1

```python
T = (s.period)/64
```

```python
Samps = []
n = []
for k in range(64):
    t = k*T
    n.append(t)
    num = s.evaluate(t) + 1.65
    num = num*((math.pow(2,12))/3.3) - 1
    Samps.append(num)
```

```python
plot.stem(n, Samps)
```

```python
f = open("DacSamps.txt","a")
for k in Samps:
    i = int(k)
    f.write(str(i))
    f.write(",\n")
```
